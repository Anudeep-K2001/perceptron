```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
import math
```

```
def rint(start = -10, end = 10):
  return random.randint(start, end)
```

```
#Generating data


def generate_data(count = 150):

  X = np.random.random(count)
  Y = np.random.random(count)
  Z = np.random.random(count)

  h = rint()
  k = rint()
  l = rint()

  r1 = (X - h)**2 + (Y - k)**2 + (Z - l)**2
  r2 = (X - h - rint())**2 + (Y - k - rint())**2 + (Z - l - rint())**2

  set1 = np.concatenate((X.reshape(count,1), r1.reshape(count,1), np.full((count,1), 0)), axis = 1)
  set2 = np.concatenate((X.reshape(count,1), r2.reshape(count,1), np.full((count,1), 1)), axis = 1)

  print()

  final_set = np.concatenate((set1, set2))

  return final_set
```

```
w = np.array([1.,1.])
```

```
def activation(val):
  if val > 0:
    return 1
  else:
    return 0
```

```
def model(inp, target, pred = 0):
  if not pred:
    inp = np.array(inp)
    for i in range(len(inp)):
      calc = np.dot(inp[i], w.T)
      if activation(calc) == target[i]:
        print(f"Expected {target[i]}\tOutput {activation(calc)}")
        #continue
      else:
        for j in range(len(w)):
          if calc > 0:
            w[j] *= 0.75                           # here 0.75 is learning rate
          else:
            w[j] /= 0.75
        print(f"Expected {target[i]}\tOutput {activation(calc)}\tweights {w}")
  if pred:
    print(f"output {activation(np.dot(inp,w.T))}")
```

```
data = pd.DataFrame(generate_data())
X = data.iloc[:,0]
Y = data.iloc[:,1]
target = data.iloc[:,2]
inp = pd.concat((X,Y), axis = 1)
```

```
model(inp, target)
```
